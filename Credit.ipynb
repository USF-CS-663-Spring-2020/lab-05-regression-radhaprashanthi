```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Read the data
df = pd.read_csv('Credit.csv')
df
```

### Income vs Limit

```python
sns.scatterplot(x='Income', y='Limit', data=df)
```

```python
sns.residplot('Income', 'Limit', data=df)

#random distribution around the regression line. So good
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(df[['Limit']], df[['Income']])
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```

#### Income can be a predictor for Limit using SLR for the line y = 0.012 * x - 12.056

```python
from sklearn.model_selection import train_test_split

X = df['Income'].values.reshape(-1,1)
y = df['Limit'].values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```

```python
y_pred = lin_reg.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.show()
```

The above graph shows SLR can be used to predict Limit based on Income
